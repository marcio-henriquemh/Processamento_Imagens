**Histograma de Equalização**

```python
#Bibliotecas

import numpy as np
from skimage import io
import matplotlib.pyplot as plt
import glob
import matplotlib.image as mpimg
import torch
import os
import cv2
import pandas as pd
from skimage.io import imread
from pathlib import Path
import math

imagens_preco =cv2.imread('/content/drive/MyDrive/Atividade-pontuada-Processa_imagens/precos/img0001.png',0)
print(imagens_preco)


def histograma(imagens_preco,nbins):
    # obtem o tamanho da matriz
    linha, coluna = imagens_preco.shape
    # cria histograma vazio, com 'nbins' posicoes
    hist = np.zeros(nbins).astype('uint8')
 

    #percorre imagem, contando valores
    for x in range(linha):
        for y in range(coluna):
            hist[imagens_preco[x,y]] += 1 # incrementa histograma no valor A[x,y]
            
    return(hist)
```

```python
hist_A = histogram(imagens_preco,256)
print(hist_A)
```

```python
plt.bar(range(256), hist_A)
plt.xlabel('Valor de intensidade')
plt.ylabel('Frequencia')
```

```python
def equaliza_histo(imagens_preco, values):
    # obtem histograma em matriz com total de valores dado por 'values'
    hist = histograma(imagens_preco, values)
    
    # cria histograma cumulativo vazio, do mesmo tamanho do histograma 
    histC = np.zeros(hist.shape).astype(int)

    # calcula histograma cumulativo
    histC[0] = hist[0]
    for i in range(1, values):
        histC[i] = hist[i] + histC[i-1]

    # tamanho da imagem/matriz    
    linha, coluna = imagens_preco.shape
    
    # o vetor abaixo e usado para guardar a funcao de transformacao, 
    # apenas para propositos didaticos (iremos visualizar posteriormente)
    hist_transform = np.zeros(values).astype(np.uint8)
    
    # nova imagem equalizada
    imagem_preco_eq = np.zeros(imagens_preco.shape).astype(np.uint8)
    
    # transforma as intensidades de 'imagem_preco' usando histC como funcao de transferencia
    # para cada tom de cinza
    for r in range(values):
        # computa qual seria o novo valor 's', dado o valor atual 'r' 
        s = ((values-1)/float(coluna*linha))*histC[r]
        
        # em toda posicao de "imagem_preco" que existir um valor 'r', 
        # entao na nova imagem esse pixel sera 's'
        imagem_preco_eq[ np.where(imagens_preco == r) ] = s
        
        # guardamos a transformacao
        hist_transform[r] = s
    
    return (imagem_preco_eq, hist_transform)
```

```python
img1_eq, img1_transf = equaliza_histo(imagens_preco, values=256)
```

```python
# novos histogramas
histeq_img1 = histograma(img1_eq, 256)
```
