Após desenvolver a aplicação em Python e Qt Creator, vamos pegar as 3 imagens geradas e verificar seus respectivos histogramas.

```python
#Bibliotecas
import numpy as np
from skimage import io
import matplotlib.pyplot as plt
from scipy.fft import dct
```

**IMAGEM ORIGINAL**

```python
img_original = io.imread ('/content/drive/MyDrive/Atividade_pontuada3/imagens/imagem_original.png')
# construcao do histograma da imagem
h = np.histogram (img_original.ravel(), bins=256, range=(0,255))
```

```python
# Visualizacao da imagem e de seu histogram
plt.imshow (img_original,cmap='gray',vmin=0, vmax=255)
plt.show()
plt.plot (h[0],'-k')
plt.show()
```

**IMAGEM COM FILTRO GAUSSIANO QUADRADO**

```python
img_gaussiana = io.imread ('/content/drive/MyDrive/Atividade_pontuada3/imagens/imagem_gaussiana.png')
# construcao do histograma da imagem
h = np.histogram (img_gaussiana.ravel(), bins=256, range=(0,255))
```

```python
# Visualizacao da imagem e de seu histogram
plt.imshow (img_gaussiana,cmap='gray',vmin=0, vmax=255)
plt.show()
plt.plot (h[0],'-k')
plt.show()
```

**IMAGEM GAUSSIANA COM CONVOLUÇÃO**

```python
img_gaussiana_convolucao = io.imread ('/content/drive/MyDrive/Atividade_pontuada3/imagens/imagem_gaussiana_convolucao.png')
# construcao do histograma da imagem
h = np.histogram (img_gaussiana.ravel(), bins=256, range=(0,255))
```

```python
# Visualizacao da imagem e de seu histogram
plt.imshow (img_gaussiana_convolucao,cmap='gray',vmin=0, vmax=255)
plt.show()
plt.plot (h[0],'-k')
plt.show()
```

**Equalizacao Histogramica** 

Imagem Original

```python

_,img_ori = plt.subplots (1,2)
plt.title("Imagem Original")
img_ori[0].imshow (img_original,cmap='gray',vmin=0,vmax=255)
h = np.histogram (img_original,bins=256,range=(0,256))
img_ori[1].plot (h[0],'-k')

_,img_gauss = plt.subplots (1,2)
plt.title("Imagem Gaussiana")
img_gauss[0].imshow (img_gaussiana,cmap='gray',vmin=0,vmax=255)
h = np.histogram (img_gaussiana,bins=256,range=(0,256))
img_gauss[1].plot (h[0],'-k')


_,img_gauss_conv = plt.subplots (1,2)
plt.title("Imagem Gaussiana convolução")
img_gauss_conv[0].imshow (img_gaussiana_convolucao,cmap='gray',vmin=0,vmax=255)
h = np.histogram (img_gaussiana_convolucao,bins=256,range=(0,256))
img_gauss_conv[1].plot (h[0],'-k')
```

**Stretch do histograma**

```python
original_img_1 = img_original - img_original.min()
original_img_1 = (original_img_1 / original_img_1.max() * 255).astype ('uint8')
_,img_ori = plt.subplots (1,2)
plt.title("Imagem Original")
img_ori[0].imshow (original_img_1,cmap='gray',vmin=0,vmax=255)
h = np.histogram (original_img_1,bins=256,range=(0,256))
img_ori[1].plot (h[0],'-k')


gauss_img_2 = img_gaussiana - img_gaussiana.min()
gauss_img_2 = (gauss_img_2 / gauss_img_2.max() * 255).astype ('uint8')
_,img_gauss = plt.subplots (1,2)
plt.title("Imagem Gaussiana")
img_gauss[0].imshow (gauss_img_2,cmap='gray',vmin=0,vmax=255)
h = np.histogram (original_img_1,bins=256,range=(0,256))
img_gauss[1].plot (h[0],'-k')


gauss_conv_img_3 = img_gaussiana_convolucao - img_gaussiana_convolucao.min()
gauss_conv_img_3 = (gauss_conv_img_3 / gauss_conv_img_3.max() * 255).astype ('uint8')
_,img_gauss_conv = plt.subplots (1,2)
plt.title("Imagem Gaussiana Convolução")
img_gauss_conv[0].imshow (gauss_conv_img_3,cmap='gray',vmin=0,vmax=255)
h = np.histogram (gauss_conv_img_3,bins=256,range=(0,256))
img_gauss_conv[1].plot (h[0],'-k')
```

```python
bac_img_original = img_original.copy()
bac_img_original[0,0] = 0
bac_img_original[0,1] = 255
bac_img_original = bac_img_original - bac_img_original.min()
bac_img_original = (bac_img_original / bac_img_original.max() * 255).astype ('uint8')
_,img_ori = plt.subplots (1,2)
plt.title("Imagem Original")
img_ori[0].imshow (bac_img_original,cmap='gray',vmin=0,vmax=255)
h = np.histogram (bac_img_original,bins=256,range=(0,256))
img_ori[1].plot (h[0],'-k')




bac_img_gaussina = img_gaussiana.copy()
bac_img_gaussina[0,0] = 0
bac_img_gaussina[0,1] = 255
bac_img_gaussina = bac_img_gaussina - bac_img_gaussina.min()
bac_img_gaussina = (bac_img_gaussina / bac_img_gaussina.max() * 255).astype ('uint8')
_,img_gauss = plt.subplots (1,2)
plt.title("Imagem Gaussiana")
img_gauss[0].imshow (bac_img_gaussina,cmap='gray',vmin=0,vmax=255)
h = np.histogram (bac_img_gaussina,bins=256,range=(0,256))
img_gauss[1].plot (h[0],'-k')


bac_img_gaussina_convolucao = img_gaussiana_convolucao.copy()
bac_img_gaussina_convolucao[0,0] = 0
bac_img_gaussina_convolucao[0,1] = 255
bac_img_gaussina_convolucao = bac_img_gaussina_convolucao - bac_img_gaussina_convolucao.min()
bac_img_gaussina_convolucao = (bac_img_gaussina_convolucao / bac_img_gaussina_convolucao.max() * 255).astype ('uint8')
_,img_gauss_conv = plt.subplots (1,2)
plt.title("Imagem Gaussiana Convolução")
img_gauss_conv[0].imshow (bac_img_gaussina_convolucao,cmap='gray',vmin=0,vmax=255)
h = np.histogram (bac_img_gaussina_convolucao,bins=256,range=(0,256))
img_gauss_conv[1].plot (h[0],'-k')
```

```python
# definicao da funcao de transferencia

# integral do histograma da imagem de entrada
l,c,a = bac_img_original.shape # numero de linhas e colunas da imagem de entrada
n = l*c # numero de pixels
pr = np.histogram (bac_img_original,bins=256,range=(0,256))[0] / n # funcao de massa de probabilidade
T = np.cumsum (pr) # funcao de transferencia
plt.title("Imagem Original")
plt.plot (T,'-k')


```
